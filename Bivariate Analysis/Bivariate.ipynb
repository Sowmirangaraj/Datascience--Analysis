```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
calc_vif(dataset[["etest_p", "salary"]])
```

ANAVO Analysis of Variance

One way Classification

```python
import scipy.stats as stats

stats.f_oneway(dataset['ssc_p'],dataset['hsc_p'],dataset['degree_p'])
```

Two way Classification

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Example:Dataset contains:
# - Numeric columns: 'ssc_p', 'hsc_p', 'degree_p', 'etest_p', 'mba_p'
# - Categorical columns: 'gender', 'degree_t'

# List of numeric columns to analyze
score_columns = ['ssc_p', 'hsc_p', 'degree_p', 'etest_p', 'mba_p']

# Loop through each score and perform two-way ANOVA
for score in score_columns:
    print(f"\n=== Two-Way ANOVA for: {score} ===")
    
    # Build and fit the model
    model = ols(f'{score} ~ C(gender) + C(degree_t) + C(gender):C(degree_t)', data=dataset).fit()
    
    # Get ANOVA table
    anova_table = sm.stats.anova_lm(model, typ=2)
    
    print(anova_table)
```
